# Matplotlib

* Görselleştirme insan beyninin verileri anlamasını kolaylaştıran en kullanışlı ve etkili yöntemlerden bir tanesidir.
* Veri bilimciler de görselleştirme araçlarından hem verileri  anlamöada hem de sonuçlarını paylaşmada çok sık yararlanırlar.
* Bu kütüphane, kullanım kolaylığı ve yetenekleriyle veri bilimcilerin araç setinin ayrılmaz bir parçasıdır.
* Matplotlib, birçok çizim (plot) tiplerini oluşturmaya imkan verdiği gibi esasında Python daki diğer çizim paketlerinin bir çoğunun da temellerini oluşturmaktadır
* Öncelikle kütüphanenin pyplot bölümüne odaklanmalıyız


```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.style.use('seaborn-darkgrid') # style vermek için
```

```python
#line chart oluşturalım
x= np.array([1,2,3,4,5,6])
y=np.power(x,3)
plt.plot(x,y)
plt.show
```

```python
plt.figure(figsize=(10,5))
x= np.linspace(0,10,1000) #sine graph çizdirme
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize= (10,5))
x= np.linspace(0,10,40)
y=np.sin(x)
plt.plot(x,y,'ro') #red green blue için renkleri seçip + yaparsan büyük o yaparsan noktalı -- çizgili yapıyor

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
plt.figure(figsize= (10,5))
x= np.linspace(0,10,40)
y=np.sin(x)
plt.plot(x,y,'rv')  # v veya ^kullanarak üçgen yaapbiliriz

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
plt.figure(figsize=(10,5))
x=np.array([1,2,3,4,5,6])
y1=np.power(x,2)
y2=np.power(x,3)
plt.plot(x,y1,'-r^',label= "$y1 = x^2$")
plt.plot(x,y2, 'g--',label="$y2 = x^3$")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.tight_layout()#constrat_layout da  adjust etmemize yardımcı oluyor
plt.show
```

```python
plt.figure(figsize=(10,6))
x= np.linspace(0,10,2000)
plt.plot(x,np.sin(x),label= "$ySin(X)$")
plt.plot(x,np.cos(x),label="$COS(X)$")
         
#plt.fill_between(x,0,np.sin(x)) 
#plt.fill_between(x,0,np.cos(x))
plt.xlabel(r'$x$',fontsize=18)
plt.ylabel(r'$Y$',fontsize=18)
plt.title("$Sin(x) $ $ & $ $ Cos(x)$",fontsize=14)
plt.legend(loc= 'lower left')
plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(14,6))
x= np.linspace(0,10,100)
y1= np.sin(x)
y2=np.cos(x)
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.show()

```

```python
#scatter graphs
x1=np.array([250,150,350,252,450,550,455,358,158,355])
x2=np.array([40,50,80,90,100,50,60,80,54,45])

y1=np.array([100,100,300,220,400,500,450,380,180,350])
y2= np.array([400,500,800,900,1000,500,600,808,504,405])
#graph 1
plt.scatter(x1,y1)
plt.xlabel("$Time$ $Spents $ ",fontsize= 12)
plt.ylabel("$Score$", fontsize=12)
plt.title("Scatter Graph")
plt.show()
```

```python
#graph 2
plt.scatter(x1,y1,label= "Class 1",marker='o',color='m')
plt.scatter(x2,y2,label="Class 20",marker= 'v',color= 'r')

plt.xlabel("$Time$ $Spents $ ",fontsize= 12)
plt.ylabel("$Score$", fontsize=12)
plt.title("Scatter Graph")
plt.legend()
plt.show()
```

```python
#pie charts
mpl.rcParams.update(mpl.rcParamsDefault)
%matplotlib inline
plt.style.use('ggplot')
plt.figure(figsize=(9,9))
area= [48,30,20,15]
labels=["Low","Medium","High","Critical"]
colors= ["#8BC34C","#D4E157","#FFB300","#FF7043"]
plt.pie(area, labels=labels, colors= colors, startangle=45)
plt.show()
```

```python
plt.figure(figsize=(8,8))
area= [48,30,20,15]
labels=["Low","Medium","High","Critical"]
colors= ["#8BC34C","#D4E157","#FFB300","#FF7043"]
plt.pie(area, labels=labels, colors= colors, startangle=45,shadow= True, autopct='%1.1f%%', explode=[0,0,0,0.1])
plt.show()
```

```python
#donut plot
plt.figure(figsize=(9,9))
area= [48,30,20,15]
labels=["Low","Medium","High","Critical"]
colors= ["#8BC34C","#D4E157","#FFB300","#FF7043"]
plt.pie(area, labels=labels, colors= colors, startangle=45)
my_circle=plt.Circle((0,0),0.7,color='white') #merkeze daire ekleme
p=plt.gcf() #gca-gcf özel fonksiyon
p.gca().add_artist(my_circle)
plt.show()
```

# Histogram Hesaplama
* sayısal verilerin dağılımının görsel bir gösterimidir.
* Karl Pearson bu konsepti bir asırdan fazla bir süre önce tanıtmışıtr.
* Histogram, sürekli veriler için kullanılan bir tü rçubuk grafiktir, çubuk grafik ise görsel olarak kategorik değişkenleri temsil eder.
* İlk adım olarak , tüm değer aralığınıızı bir dizi aralığa bölmek gerekir. Her kutu bitişik olmalıdır ve hiçbiri üst üste gelemez. G
* Genel olarak, kutu boyları eşittir ve dahil edilecek kutu sayısınnın adedi 5-20 dir.
* Bunun anlamı eğer 20 den fazla kutuya sahipseniz, grafiğinizin okumak zor olacaktır. Aksine 5ten az olursa da grafiğiniz verilerinizin dağılımını hakkında yeterince bilgi vermeyecektir.

```python
#histogram

x= np.random.rand(2000)
plt.hist(x,bins = 40, color= "purple")
plt.gca().set(title= "Histogram", ylabel= "Frequency")
plt.show()
```

```python
x= np.random.rand(2000)
plt.hist(x,bins = 30, color= "green")
plt.gca().set(title= "Histogram", ylabel= "Frequency")
plt.show()
```

```python
plt.figure(figsize= (10,8))
x= np.random.rand(2000)
plt.hist(x,bins = 30, color= "green", edgecolor= "red")
plt.gca().set(title= "Histogram", ylabel= "Frequency")
plt.show()
```

```python
x= np.random.rand(2000)
plt.hist(x,bins = 30, color= "green",edgecolor= "blue")
plt.gca().set(title= "Histogram", ylabel= "Frequency")
plt.show()
x= np.random.rand(2000)
plt.hist(x,bins = 20, color= "green",edgecolor= "blue")
plt.gca().set(title= "Histogram", ylabel= "Frequency")
plt.show()
x= np.random.rand(2000)
plt.hist(x,bins = 10, color= "green",edgecolor= "blue")
plt.gca().set(title= "Histogram", ylabel= "Frequency")
plt.show()
```

```python
plt.figure(figsize= (8,11))
x= np.random.normal(-4,1,size=800)
y=np.random.normal(-4,1,size= 800)
z=np.random.normal(3.5,1,size=800)
plt.hist(x,bins = 30, color= "green", alpha=0.6)
plt.hist(y,bins = 30, color= "purple", alpha=0.6)
plt.hist(z,bins = 30, color= "red", alpha=0.6)
plt.gca().set(title= "Histogram", ylabel= "Frequency")
plt.show()
```
